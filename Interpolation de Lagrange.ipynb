```python
from math import factorial
import numpy as np
import numpy.linalg as LA
import matplotlib.pyplot as plt
```

```python
# Construction d'interpolation de Lagrange par la methode direct
# P(Xi) = A0 + A1.X1 + A2.X2 + An.Xn
# Supposons les points suivants : A1(1, 1.4), A2(1.4, 3), A3(2, 3.5)
# Si n = 3, nous avons alors un polynome de degre 2

def Interpolation(P0=[], P1=[], P2=[]):

    P0 = np.array(P0)
    P1 = np.array(P1)
    P2 = np.array(P2)
    P3 = np.array([])
    
    Pol_X0 = np.array([1, P0[0], P0[0]**2], P0[0]**3)
    Pol_X1 = np.array([1, P1[0], P1[0]**2], P1[0]**3)
    Pol_X2 = np.array([1, P2[0], P2[0]**2], P2[0]**3)
#     Pol_X3 = np.array([1, P2[0], P2[0]**2], P3[0]**3)
    
    Pol_X = np.stack((Pol_X0, Pol_X1, Pol_X2))
    Yi = np.array([P0[1], P1[1], P2[1]])
    Lk = LA.solve(Pol_X, Yi)
#         print('Une erreur s\'est produite, le determinant est nulle')
#         exit()
        
    # Lk(X) = -7.0333 + 11.6*X -3.1666X**2
    Xi = np.linspace(-1, 5, 20)
    Yi = Lk[0] + Lk[1]*Xi + Lk[2]*(Xi**2)
#     print(f'Lk = {Lk[0]} + {Lk[1]}.Xk + {Lk[2]}.Xk^2')
    
    plt.plot(Xi, Yi)

    # Tracage des points P(Xi) = f(Xi) 
    plt.scatter(P0[0], P0[1], color='black')
    plt.scatter(P1[0], P1[1], color='black')
    plt.scatter(P2[0], P2[1], color='black')
```
